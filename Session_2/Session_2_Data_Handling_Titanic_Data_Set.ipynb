```python
#Distrubition of age-fare (Swarm Plot)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=[12,10])
sns.swarmplot(x="Sex",y="Fare",data=df)
plt.show()
```

```python


sns.lmplot('Age',#horizontal axis
          'Fare',#vertical axis
          data=df, #data source
          fit_reg=False, #don't fit a reg file
          hue="Survived", #set a color
          scatter_kws={"marker":"D",#set marker style
                       "s": 100,}) # marker size}
#Set a Title
plt.title("Histogram of IQ")
#Set x-axis label
plt.xlabel("Age")
#Set y-axis label
plt.ylabel("Fare")
plt.show()                      
                         
```
