```python
import pandas as pd
import numpy as np
df= pd.read_csv("C:/Users/Ecem/Desktop/BTS/BTS_MasterInBigData-master/Session_2/2_titanic_dataset.csv")
df.head()
```

```python
####CountPlot#####


import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=[12,10]) 
sns.countplot(df['Survived']) 
plt.show()
```

```python
##Bar Plot###

plt.figure(figsize=[12,10])
sns.barplot('Sex','Survived',data=df)
plt.show()
```

```python
###Point Plot####
sns.pointplot(x="Parch",y="Survived",hue="Sex",data=df,
             palette={"male":"blue","female":"red"})
plt.show()
```

```python
#######Swarm Plot######
plt.figure(figsize=[12,10])
sns.swarmplot(x="Sex",y="Age",data=df)
plt.show()
```

```python
####factor Plot#####
plt.figure(figsize=[12,10])
sns.factorplot("Pclass","Survived","Sex",data=df,kind="bar",palette="muted",legend=False)
plt.show()
```

```python
####Scatter_Plot#######

sns.lmplot('Age',#horizontal axis
          'Fare',#vertical axis
          data=df, #data source
          fit_reg=False, #don't fit a reg file
          hue="Sex", #set a color
          scatter_kws={"marker":"D",#set marker style
                       "s": 100,}) # marker size}
#Set a Title
plt.title("Histogram of IQ")
#Set x-axis label
plt.xlabel("Age")
#Set y-axis label
plt.ylabel("Fare")
plt.show()                      
                       
```

```python
#Scatter Plots

g=sns.FacetGrid(df,col="Sex",row="Pclass",margin_titles=True)
g.map(plt.scatter,"Fare","Age")

#Add a title to the figure
g.fig.suptitle("this is a title")

#Show the plot
plt.show()
```

```python
######Distplot########

g=sns.FacetGrid(df,col="Sex")
g.map(sns.distplot, "Fare")
plt.show()
```
